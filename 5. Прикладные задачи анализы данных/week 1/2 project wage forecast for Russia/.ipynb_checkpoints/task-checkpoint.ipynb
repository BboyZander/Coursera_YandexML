```python
from __future__ import division

import numpy as np
import pandas as pd

from scipy import stats
import statsmodels.api as sm

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter('ignore')


from itertools import product
from datetime import *
from dateutil.relativedelta import *

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

%matplotlib inline
```

```python
#Reading data
wage = pd.read_csv('WAG_C_M.csv', sep=';', index_col='month', parse_dates=True, dayfirst=True)
wage.info()
```

```python
wage.head()
```

## Визуальный анализ ряда

```python
wage.plot(figsize=(12, 9), grid=0.2)
plt.title('Среднемесячный уровень заработной платы')
```

свойства ряда:
* повышающийся тренд
* годовая сезонность
* автокоррелированность
* видно, что с течением времени увеличивается десперсия -> ряд не является стационарным

```python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.WAG_C_M)[1])
_ = sm.tsa.seasonal_decompose(wage.WAG_C_M).plot()
```

## Стабилизация дисперсии

Сезонный профиль достаточно четко выражен

```python
wage['WAG_C_M_box'], lmbda = stats.boxcox(wage.WAG_C_M)

_ = plt.figure(figsize=(15,7))
_ = wage.WAG_C_M_box.plot()
_ = plt.title(u'Стабилизация дисперсии')
_ = plt.grid(alpha=0.2)

print('Оптимальный параметр преобразования Бокса-Кокса: %f' % lmbda)
print('Критерий Дики-Фуллера p=%f' % sm.tsa.stattools.adfuller(wage.WAG_C_M_box)[1])
```

Ряд по прежнему не является стационарным. Применим сезонное дифференцирование

```python
wage['WAG_C_M_box_diff'] = wage.WAG_C_M_box - wage.WAG_C_M_box.shift(12)
_ = sm.tsa.seasonal_decompose(wage.WAG_C_M_box_diff.dropna()).plot()
_ = plt.grid(alpha=0.2)
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.WAG_C_M_box_diff[12:])[1])
```

Согласно критерию дики_фуллера гипотеза о нестационарности отвергается на уровне значимости 0.05. Однако на картинка все еще виден тренд. Применим к текущим данным обычное дифференцирование

```python
wage['WAG_C_M_box_diff2'] = wage.WAG_C_M_box_diff - wage.WAG_C_M_box_diff.shift(1)
_ = sm.tsa.seasonal_decompose(wage.WAG_C_M_box_diff2.dropna()).plot()
print('Критерий Дики-Фуллера p=%f' % sm.tsa.stattools.adfuller(wage.WAG_C_M_box_diff2.dropna())[1])
```

Теперь можем считать, что ряд стационарен

## Выбор начальных приблежений для p, q, P, Q

Посмотрим на ACF и PACF полученного ряда:

```python
plt.figure(figsize=(15,10))

ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(wage.WAG_C_M_box_diff2.dropna()[12:].squeeze(), lags=50, ax=ax);
ax.grid(alpha=0.2)

ax = plt.subplot(212)
ax.grid(alpha=0.2)
sm.graphics.tsa.plot_pacf(wage.WAG_C_M_box_diff2.dropna()[12:].squeeze(), lags=50, ax=ax);
```

```python
# Начальные приближения: Q=0, q=1, P=1, p=1

ps = range(0, 2)
d=1
qs = range(0, 2)
Ps = range(0, 2)
D=1
Qs = range(0, 1)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

## Обучение и сравнение моделей-кандидатов, выбор победителя

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(wage.WAG_C_M_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
print(best_model.summary())
```

## Анализ остатков

```python
_ = plt.figure(figsize=(15,12))
_ = plt.subplot(211)
_ = best_model.resid[13:].plot()
_ = plt.ylabel(u'Residuals')

_ = ax = plt.subplot(212)
_ = sm.graphics.tsa.plot_acf(best_model.resid.values.squeeze(), lags=50, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Посмотрим, насколько хорошо модель описывает данные:

```python
def invboxcox(y,lmbda):
    if lmbda == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lmbda*y+1)/lmbda))
    
wage['model'] = invboxcox(best_model.fittedvalues, lmbda)

_ = plt.figure(figsize=(15,7))
_ = wage.WAG_C_M.plot()
_ = wage.model[13:].plot(color='r')
_ = plt.title('Average nominal wage')
_ = plt.grid(alpha=0.2)
```

## Прогнозирование с помощью финальной модели.

```python
wage2 = wage[['WAG_C_M']]
date_list = [datetime.strptime("2017-07-01", "%Y-%m-%d") + relativedelta(months=x) for x in range(0,36)]
future = pd.DataFrame(index=date_list, columns=wage2.columns)
wage2 = pd.concat([wage2, future])
wage2['forecast'] = invboxcox(best_model.predict(start=294, end=329), lmbda)

_ = plt.figure(figsize=(15,7))
_ = wage2.WAG_C_M.plot(label = 'history')
_ = wage2.forecast.plot(color='r', label='predicted')
_ = plt.legend()
_ = plt.title('Среднемесячный уровень заработной платы')
_ = plt.grid(alpha=0.2)
```

Прогноз выглядит адекватным, соответствует тренду и передает информацию о сезонности
